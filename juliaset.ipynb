```python
import cmath;

class JuliaSet(object):
  
    def __init__(self, cReal, cImg, n=100, _d=0.001):
        self.c=complex(cReal, cImg);
        self._d=_d;
        if(n>0):
            self.n=n;
        else:
            print "Reset n to 100";
       
        
        
    def juliamap(self, z):
        return self.c+(z**2);
    
    
    def iterate(self, z):
        m=0;
        while(1>0):
            z=self.juliamap(z);
            m+=1;
            if(abs(z)>2):
                return m;
            elif(m>=self.n):
                return 0;
    
    
    def setcomplexplane(self, d=0):
        i =-2;
        if(d>0):
            increment = d;
        else:
            increment = self._d;
              
        self._complexplane = [];
        while(i<=2):
            x,y=i,i;
            z = complex(x,y);
            self._complexplane.append(z);
            i+=increment;    
        
        
    def set_spacing(self, d):
        setcomplexplane(self, d);
        
    
    def generate(self):
        a = [];
        for v in self._complexplane:
         a.append(self.iterate(v));   
        self.set=a;
        return a;
```

```python
js=JuliaSet(0.5,0.5,5,0.1);
js.setcomplexplane();
print js.generate();
```
